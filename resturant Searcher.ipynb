# For Swiggy

```python
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException

#connect to chrome app
driver = webdriver.Chrome()
driver.get("https://www.swiggy.com/city/bangalore")
```

```python
#click on search button
search = driver.find_element(By.XPATH, '(//button)[1]')
search.click()

#search needed food item
input = driver.find_element(By.XPATH, '//input')
input.send_keys('Pani Puri'+ Keys.ENTER)
sleep(3)

#click on resturant button of swiggy
resturant_button = driver.find_element(By.XPATH, '//span[@class="styles_container__1ieVH styles_size-sm__1emwv styles_variant-default__2VzW8 styles_rounded__3IerS styles_canClick__16iph NavTab_tab__1JygM"]')
resturant_button.click()

print('setup is done')
```

```python
#in swiggy getting the resturant name, raitings, prices for 2
swiggy_resturant_names=[]
swiggy_resturant_raitings=[]
swiggy_prices_for_two=[]
flag = 1
i = 0
try:
 while flag == 1:
    resturant_name = driver.find_element(By.XPATH, f'(//div[@class="styles_restaurantName__29jAP"])[{i+1}]')
    swiggy_resturant_names.append(resturant_name.text)

    resturant_raiting = driver.find_element(By.XPATH, f'(//span[@class="styles_restaurantMetaRating__4H1gt"])[{i+1}]')
    swiggy_resturant_raitings.append(resturant_raiting.text)

    price_for_two = driver.find_element(By.XPATH, f'(//div[@data-testid="restaurant-card-cost"])[{i+1}]')
    swiggy_prices_for_two.append(price_for_two.text)
    i+=1
except NoSuchElementException:
    flag = 0


```

```python
swiggy_data = swiggy_resturant_names, swiggy_resturant_raitings, swiggy_prices_for_two
swiggy_data
```

```python
import numpy as np
swiggy_data_numpy = np.array(swiggy_data)
swiggy_data_numpy = swiggy_data_numpy.transpose()
swiggy_data_numpy.shape
```

# Zomato starts here

```python
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException


driver = webdriver.Chrome()
driver.get("https://www.zomato.com/bangalore/delivery")
```

```python
# click on search button
search = driver.find_element(By.XPATH, '//input[@class="sc-fxgLge jUPfKP"]')
search.click()

#add your food item
input = driver.find_element(By.XPATH, '//input[@class="sc-fxgLge jUPfKP"]')
input.send_keys('Pani puri'+ Keys.ENTER)
sleep(3)

#click on the delivery dish
delivery_items = driver.find_element(By.XPATH, '(//div[@class="sc-imAxmJ sc-iWadT bImqII"])[1]')
delivery_items.click()

#click on 4+ raiting
raiting_button = driver.find_element(By.XPATH, '//div[@class="sc-dYzWWc rQRkV pill_rating_range"]')
raiting_button.click()

print('Wait patiently as zomato updates new cards while scrolling so selenium is scrolling through the entire page to get all the cards loaded to move forward.')
#scroll the entire page to load all the possible cards
for i in range (600):
    driver.find_element(By.TAG_NAME, 'Body').send_keys(Keys.ARROW_DOWN)

print('setup is done')
```

```python
#in zomato getting the resturant name, raitings, prices for 1
zomato_resturant_names=[]
zomato_resturant_raitings=[]
zomato_prices_for_one=[]
flag = 1
i = 0
try:
 while flag == 1:
    resturant_name = driver.find_element(By.XPATH, f'(//h4[@class="sc-1hp8d8a-0 sc-cgThhu dFwWJC"])[{i+1}]')
    zomato_resturant_names.append(resturant_name.text)

    resturant_raiting = driver.find_element(By.XPATH, f'(//div[@class="sc-1q7bklc-1 cILgox"])[{i+1}]')
    zomato_resturant_raitings.append(resturant_raiting.text)

    price_for_two = driver.find_element(By.XPATH, f'(//p[@class="sc-1hez2tp-0 sc-LAuEU dHtbEm"])[{i+1}]')
    zomato_prices_for_one.append(price_for_two.text)
    i+=1
except NoSuchElementException:
    flag = 0
```

```python
zomato_data = zomato_resturant_names, zomato_resturant_raitings, zomato_prices_for_one
zomato_data
```

```python
import numpy as np
zomato_data_numpy = np.array(zomato_data)
zomato_data_numpy = zomato_data_numpy.transpose()
zomato_data_numpy.shape
```

# making dataframes for futher opertions

```python
import pandas as pd
swiggy_df = pd.DataFrame(swiggy_data_numpy, columns = ['Resturant Name', 'Resturant Raitings', 'Resturant Price'])
swiggy_df
```

```python
zomato_df = pd.DataFrame(zomato_data_numpy, columns = ['Resturant Name', 'Resturant Raitings', 'Resturant Price'])
zomato_df
```

```python
new_df = pd.concat([swiggy_df, zomato_df], ignore_index=True)
new_df
```
